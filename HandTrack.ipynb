```python
import cv2
import mediapipe as mp
import time
```

```python
cap = cv2.VideoCapture(0) #for camera selection

#making hand selection point

mpHands = mp.solutions.hands # for hand selection
hands = mpHands.Hands() #for enabling the library
mpDraw = mp.solutions.drawing_utils # finally draw the fucnction

#fps
pTime=0 # previous time 0 te set koralm
cTime=0 #current time 0 te set koralm

while True:
    success, img = cap.read() #camera theke input ta nebe
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #jupyter BGR khub pochondo kore tai o default e bgr e rakhe otake amra RGB te kore nilam noile colour ultopalta asto
    result = hands.process(imgRGB) #result er modheye amra o/p ta store korbo
    #print(result.multi_hand_landmarks) <-- ekhane screen er upor e kichu na thakle none print koreb r hat anle value debe
    
    if result.multi_hand_landmarks: #jodi result e multi hand thake
        for hand_landmarks in result.multi_hand_landmarks: # tale ami akta loop chalabo jeta 2 to hat kei nebe r proper kaj korbe function diye
            for id,lm in enumerate(hand_landmarks.landmark): # to get the value of every point for hand
                # print(id, lm) #lm = landmarks
                h, w, c = img.shape
                cx, cy = int(lm.x * w), int(lm.y * h)
                print(id, cx, cy)
                
                #if id == 0:
                #   cv2.circle(img, (cx, cy), 15, (255,0,255), cv2.FILLED)
                
            mpDraw.draw_landmarks(img, hand_landmarks, mpHands.HAND_CONNECTIONS) #o/p e oi line guloo mane main output ta prow kora holo
   
    #fps er jonne
    cTime = time.time()
    fps = 1 / (cTime - pTime)
    pTime = cTime
    
    cv2.putText(img, str(int(fps)), (10,70), cv2.FONT_HERSHEY_PLAIN, 3, (255,0,255), 3)#fps print korbe
    
        
        
    cv2.imshow("image", img) #output er main box ta dekhacche
    cv2.waitKey(1) 
```
